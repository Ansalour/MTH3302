# Exercice 1

```julia
# Chargement des données
data = CSV.read("bodyfat.csv")
first(data,5)
```

## a) affichage du pourcentage de graisse en fonction de l'épaisseur du pli du tricep

Dans la première cellule, on utilisera le fait que les données sont contenues dans un dataframes. Dans la deuxième cellule, on affichera le nuage de point des vecteurs exrtaits du dataframe.


```julia
# Affichage lorsque les données sont dans un dataframe
plot(data, x=:Triceps, y=:Bodyfat, Geom.point)
```

```julia
# Extraction des données dans des vecteurs

x₁ = convert(Array{Float64},data[:Triceps])
y = convert(Array{Float64},data[:Bodyfat])
plot(x=x₁, y=y, Geom.point)
```

## b) Estimez les paramètres du modèle de régression linéaire simple entre le pourcentage de graisse et l'épaisseur du pli du tricep

```julia
x₁ = convert(Array{Float64},data[:Triceps])
y = convert(Array{Float64},data[:Bodyfat])
n = length(y)

x̄₁ = mean(x₁)
ȳ = mean(y)


β̂₁ = sum( (x₁[i] - x̄₁)*(y[i] - ȳ) for i=1:n) / sum( (x₁[i] - x̄₁)^2 for i=1:n )

β̂₀ = ȳ - β̂₁*x̄₁

println("β̂₀ = $β̂₀")
println("β̂₁ = $β̂₁")
```

## c) Affichage de la droite de régression obtenue 

```julia
# Couche des données
sample = layer(data, x=:Triceps, y=:Bodyfat, Geom.point, Theme(default_color="deepskyblue"))

# droite de régression définie sur le domaine des données
xx = collect(range(minimum(x₁),stop=maximum(x₁),length=10))
regression = layer(x=xx, y = β̂₀ .+ β̂₁*xx, Geom.line, Theme(default_color="red"))

plot(sample,regression,
    Guide.manual_color_key("Légende", ["Échantillon", "Régression"], ["deepskyblue","red"]))

# Les deux premièeres hypothèses de la régression semblent semblent satisfaites
```

## d) On estime les paramètres du modèle de régression avec toutes les variables explicatives

```julia
x₁ = convert(Array{Float64},data[:Triceps])
x₂ = convert(Array{Float64},data[:Thigh])
x₃ = convert(Array{Float64},data[:Midarm])
y = convert(Array{Float64},data[:Bodyfat])
n = length(y)

X = hcat(ones(n),x₁,x₂,x₃)

β̂ = (X'X)\X'y
```

## e) Calcul et affichage des résidus du modèle

```julia
# Calcul des prévisions
ŷ = X*β̂

# Calcul des résidus
e = y-ŷ

# affichage des résidus
plot(x=ŷ, y=e, Geom.point, Coord.Cartesian(xmin=10,xmax=30), Guide.xlabel("Prévision"), Guide.ylabel("Résidu"))

# Les deux premières hypothèeses de la régression semblent satisfaites
```

# d) Difficile à dire. Il semble y avoir une relation mais très faible. Essayons quand même d'estimer le modèle de régression.

# e) Estimation du modèle de régression

```julia
x₁ = convert(Array{Float64},data[:CP1])
x₂ = convert(Array{Float64},data[:CP2])
y = convert(Array{Float64},data[:Final])
n = length(y)

# Construction d'une variable indicatrice pour la session
x₃ = zeros(n)
ind = data[:Session] .== "H2018"
x₃[ind] .= 1;

# Définition de la matrice de strucutre
X = hcat(ones(n),x₁,x₂,x₃)

# estimation des paramètres
β̂ = (X'X)\X'*y

println("β̂=$β̂")

#= Estimons la valeurs de la note au final pour un étudiant de la session H2018 
qui aurait eu 13 au CP1 et 15 au CP2=#

x₀ = [1.0, 13.0, 15.0, 1.0]

Ŷ₀ = x₀'*β̂

println("L'estimation de sa note au finale est de $Ŷ₀ sur 35.")
```
